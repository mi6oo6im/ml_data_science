- No new columns 
- No multiple rows 
- No partial rows, no changes to other rows

````python
tb_tidy = tb_data.melt(id_vars = ['iso2', 'year'], var_name = 'sex_and_age', value_name = 'cases')
````

````python
tb_tidy['gender'] = tb_tidy.sex_and_age.str.slice(0, 1)
````

````python
tb_tidy['age_group'] = tb_tidy.sex_and_age.str.slice(1)
````

````python
tb_tidy = tb_tidy.drop(columns = ['sex_and_age'])
````

````python
tb_tidy.describe().T
````

````python
tb_tidy = tb_tidy.dropna()
````

````python
tb_tidy.age_group.str.len().value_counts()
````

````python
tb_tidy.age_group.unique()
````

````python
def process_age_group(age_group):
    ages = {"04": "0-4", "65": "65+", "u": "unknown"}
    if age_group in ages:
        return ages[age_group]
    else:
        # Put a dash before the last two digits
        return f"{age_group[:-2]}-{age_group[-2:]}"

process_age_group('014')
````

````python
tb_tidy.age_group = tb_tidy.age_group.apply(process_age_group)
````
