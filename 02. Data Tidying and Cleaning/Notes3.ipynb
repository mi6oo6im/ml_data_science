```python
billboard_tidy = billboard_data.melt(id_vars = ['year', 'artist', 'track', 'time', 'date.entered'],
                   var_name = 'week',
value_name = 'position')
```

```python
billboard_tidy = bilboard_tidy.dropna()
```

```python
billboard_tidy
```

```python
billboard_tidy.dtypes
```

```python
billboard_tidy.position = billboard_tidy.position.astype(int)
```

```python
billboard_tidy
```

```python
del bilboard_tidy
```

```python
billboard_tidy['date.entered'] = pd.to_datetime(billboard_tidy['date.entered'])
```

```python
billboard_tidy.dtypes
```

```python
pd.Timedelta(1, 'w')
```

```python
intervals = billboard_tidy.week.apply(lambda x: pd.Timedelta(x - 1, 'w'))
```

```python
billboard_tidy['date'] = billboard_tidy['date.entered'] + intervals
```

```python
billboard_tidy
```

```python
billboard_tidy[billboard_tidy['track'] == 'Higher']
```

```python
billboard_tidy = billboard_tidy.drop(columns = ['date.entered', 'week'])
```

```python
billboard_tidy = billboard_tidy[['date', 'artist', 'track', 'time', 'position']]
```
